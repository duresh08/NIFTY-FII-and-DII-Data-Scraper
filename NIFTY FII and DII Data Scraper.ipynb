```python
import requests
import bs4 as BeautifulSoup

from selenium import webdriver
from selenium.webdriver import ActionChains

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.select import Select

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

import pandas as pd

import time

url = 'https://www.niftytrader.in/fii-dii-activity'

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
driver.maximize_window()

Master_df = pd.DataFrame()

Dates_list_dropdown = driver.find_elements(By.CLASS_NAME,"select-box")

Dates_options = Dates_list_dropdown[0].find_elements(By.TAG_NAME,"option")

i = 0

Dates_list = []

while i<len(Dates_options):
    Dates_list.append(Dates_options[i].text)
    i+=1
    
Dates_list = Dates_list[::-1]


for dates in Dates_list:
    Dropdown_date = driver.find_element_by_xpath('//*[@id="content"]/div[4]/div/div[1]/div[2]/div[1]/div[2]/select')
    Dropdown = Select(Dropdown_date)
    Dropdown.select_by_visible_text("{}".format(dates))
    
    time.sleep(10)
    
    list_of_dictionaries = []
    list_of_rows = driver.find_elements_by_tag_name("tr")
    
    n_rows = len(list_of_rows)
    
    i = 1
    j = 0
    
    while i in range(1,n_rows):
        
        temp_dictionary = {}
        
        cell_data = driver.find_elements_by_tag_name("td")
        
        temp_dictionary['CATEGORY'] = cell_data[j].text
        
        j+=1
        
        temp_dictionary['TRADING DATE'] = cell_data[j].text
        
        j+=1
        
        temp_dictionary['GROSS PURCHASES (RS.CRORE)'] = cell_data[j].text
        
        j+=1
        
        temp_dictionary['GROSS SALES (RS. CRORE)'] = cell_data[j].text
        
        j+=1
        
        temp_dictionary['NET INVESTMENT (RS. CRORE)'] = cell_data[j].text
        
        list_of_dictionaries.append(temp_dictionary)
        
        j+=1
        i+=1
        
    df = pd.DataFrame(list_of_dictionaries)
    Master_df = pd.concat([Master_df, df], ignore_index=True)

Master_df
```
